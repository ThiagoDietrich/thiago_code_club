```python
amostra = [
    "Melipona scutellaris",
    "Melipona quadrifasciata",
    "Melipona scutellaris",
    "Scaptotrigona postica",
    "Scaptotrigona depilis",
    "Scaptotrigona postica"
]

```

```python
for nome in amostra:
    print(nome)
```

```python
contagem_especies = {}
contagem_generos = {}

for nome in amostra:
    # separar gênero e espécie
    genero = nome.split()[0]

    # contar espécies
    if nome in contagem_especies:
        contagem_especies[nome] += 1
    else:
        contagem_especies[nome] = 1

    # contar gêneros
    if genero in contagem_generos:
        contagem_generos[genero] += 1
    else:
        contagem_generos[genero] = 1
```

```python
print("Contagem por espécie:")
for especie, qtd in contagem_especies.items():
    print(especie, ":", qtd)

print("\nContagem por gênero:")
for genero, qtd in contagem_generos.items():
    print(genero, ":", qtd)
```
